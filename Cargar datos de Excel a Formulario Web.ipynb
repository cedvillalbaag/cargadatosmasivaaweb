## Organizar código para la ejecución del script de automatización

```python
#Definir parametros
website = 'https://docs.google.com/forms/d/e/1FAIpQLSemfmim4Gwh2RS9okpCfsrAsQPWUvcVxL7tRbUa8-gsh9-WLg/viewform?usp=pp_url'
path = 'chromedriver.exe'
s = Service(path)

#Abrir explorador
driver = webdriver.Chrome(service = s)


for row, datos in df.iterrows():
    index = datos['id']
    nombre = datos['nombre']
    apellido_paterno = datos['apellido paterno']
    apellido_materno = datos['apellido materno']
    sexo = datos ['sexo']
    correo_electronico = datos['correo electronico']
    direccion = datos['direccion']
    numero_tlf = datos['numero telefono']
    comentarios = datos['comentarios']
    rubro = datos['rubro']

    #Abrir formulario
    driver.get(website)
    #Despues de abrir pagina web esperar 3s para seguir corriendo script
    time.sleep(2)   
    
    ####################################################
    #Ingreso de Informacion a Formulario (se debe sustituir los valores de las pruebas por el nombre de variables definida en ciclo):
    
    #Encontrar input id en formulario
    driver.find_element('xpath', '//*[@id="mG61Hd"]/div/div/div[2]/div[1]/div/div[2]/div/div[1]/div/div[1]/input')
    #Enviar informacion a input id
    driver.send_keys('123')
    
    #Encontrar input nombre en formulario
    driver.find_element('xpath', '//*[@id="mG61Hd"]/div/div/div[2]/div[2]/div/div[2]/div/div[1]/div/div[1]/input')
    #Enviar informacion a input nombre
    driver.send_keys(nombre)
    
    #Encontrar input apellido paterno en formulario
    driver.find_element('xpath', '//*[@id="mG61Hd"]/div/div/div[2]/div[3]/div/div[2]/div/div[1]/div/div[1]/input')
    #Enviar informacion a input apellido paterno
    driver.send_keys(apellido_paterno)
    
    #Encontrar input apellido materno en formulario
    driver.find_element('xpath', '//*[@id="mG61Hd"]/div/div/div[2]/div[4]/div/div[2]/div/div[1]/div/div[1]/input')
    #Enviar informacion a input apellido materno
    driver.send_keys(apellido_materno)
    
    #Encontrar elemento y dar click en checkbox del formulario
    driver.find_element('xpath', sexo_dict[sexo]).click()
    
    #Encontrar input correo electronico en formulario
    driver.find_element('xpath', '//*[@id="mG61Hd"]/div/div/div[2]/div[6]/div/div[2]/div/div[1]/div/div[1]/input')
    #Enviar informacion a input correo electronico
    driver.send_keys(correo_electronico)
    
    #Encontrar input direccion en formulario
    driver.find_element('xpath', '//*[@id="mG61Hd"]/div/div/div[2]/div[7]/div/div[2]/div[1]/div[2]/textarea')
    #Enviar informacion a input direccion
    driver.send_keys(direccion)
    
    #Encontrar input numero telefono en formulario
    driver.find_element('xpath', '//*[@id="mG61Hd"]/div/div/div[2]/div[8]/div/div[2]/div/div[1]/div/div[1]/input')
    #Enviar informacion a input numero telefono
    driver.send_keys(numero_tlf)
    
    #Encontrar input comentarios en formulario
    driver.find_element('xpath', '//*[@id="mG61Hd"]/div/div/div[2]/div[9]/div/div[2]/div[1]/div[2]/textarea')
    #Enviar informacion a input comentarios
    driver.send_keys(comentarios)
    
    #Abrir lista de valores desplegables (Ubicar y dar click)
    driver.find_element('xpath', '//*[@id="mG61Hd"]/div/div/div[2]/div[10]/div/div[2]/div[1]/div[1]/div[1]/span').click()
    #Despues de abrir lista, esperar 2s
    time.sleep(2)
    #Encontrar elemento y dar click en lista de formulario
    driver.find_element('xpath', rubros_dict[rubro]).click()
    time.sleep(1)
    
    
    #Agregar clic boton enviar
    driver.find_element('xpath', '//*[@id="mG61Hd"]/div/div/div[3]/div[1]/div/div/span/span').click()
```
